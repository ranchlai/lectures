```python
import numpy as np
d=32
T=40
k = np.linspace(1,d/2,d//2,dtype='float64')
t = np.linspace(0,T,T,dtype='float64',endpoint=False)
omega = 1/(10000**(2*k/d))
pos = np.empty((T,d),'float64')
pos[:,::2] = np.sin(t[:,None] @ omega[None,:])
pos[:,1::2] = np.cos(t[:,None] @ omega[None,:])
pos.shape
```

```python
# (cos(wt+w)-cos(wt))^2 + (sin(wt+w)-sin(wt))^2 
# = 2 - 2*cos(wt+w)*cos(wt) - 2*sin(wt+w)*sin(wt)
# = 2 - 2*cos(w)
```

```python
2*(1-np.cos(omega[0])),np.sum((pos[0,:2]-pos[1,:2])**2) 
```

```python
np.sum((pos[2,:]-pos[1,:])**2),np.sum(2*(1-np.cos(omega)))
```
